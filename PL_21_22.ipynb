### Step 3: Feature Engineering and Data Cleaning

#### Let's order these games by the time they were played throughout the season, and add a points column that can be tracked relative to the point in the season.

```python
# Ensuring that 'DATE_TIME' is in datetime format
football_data_cleaned['DATE_TIME'] = pd.to_datetime(football_data_cleaned['DATE_TIME'])

# Sorting the dataset in descending order of 'DATE_TIME'
football_data_sorted = football_data_cleaned.sort_values(by='DATE_TIME', ascending=True)

# Display the first few rows of the sorted dataset
football_data_sorted.head()
```

```python
# Adjust the function to calculate points
def assign_points(home_goals, away_goals):
    if home_goals > away_goals:  # Home win
        return 3, 0
    elif home_goals < away_goals:  # Away win
        return 0, 3
    else:  # Draw
        return 1, 1

# Assign points for each match
football_data_sorted['HOME_POINTS'], football_data_sorted['AWAY_POINTS'] = zip(*football_data_sorted.apply(
    lambda row: assign_points(row['HOME_SCORE'], row['AWAY_SCORE']), axis=1))

# We'll use a dictionary to keep track of the points
cumulative_points_dict = {team: [] for team in football_data_sorted['HOME'].unique()}

# Sort the matches by date
football_data_sorted = football_data_sorted.sort_values('DATE_TIME')

# Calculate the running total of points
for i, match in football_data_sorted.iterrows():
    # Add points to home team
    home_team, away_team = match['HOME'], match['AWAY']
    cumulative_points_dict[home_team].append(match['HOME_POINTS'])
    
    # If this is the first match for the away team, initialize their list of points
    if not cumulative_points_dict[away_team]:
        cumulative_points_dict[away_team] = [0] * (len(cumulative_points_dict[home_team]) - 1)
    
    # Add points to away team
    cumulative_points_dict[away_team].append(match['AWAY_POINTS'])

# To ensure all teams have the same number of games, we'll fill any missing games with the last known points
max_games = max(len(points) for points in cumulative_points_dict.values())
for team_points in cumulative_points_dict.values():
    last_known_points = team_points[-1] if team_points else 0
    team_points.extend([last_known_points] * (max_games - len(team_points)))

# Convert the dictionary into a DataFrame for easier manipulation and visualization
cumulative_points_df = pd.DataFrame(cumulative_points_dict).cumsum()

# Display the DataFrame
cumulative_points_df
```

## Let's look at Away day performances

```python
# Let's define team_away_performance by calculating the number of goals scored, goals conceded, and total points in away games.

# Calculate 'AWAY_WIN' and 'AWAY_DRAW'
football_data_cleaned['AWAY_WIN'] = (football_data_cleaned['AWAY_SCORE'] > football_data_cleaned['HOME_SCORE']).astype(int)
football_data_cleaned['AWAY_DRAW'] = (football_data_cleaned['AWAY_SCORE'] == football_data_cleaned['HOME_SCORE']).astype(int)

# Now we can create 'team_away_performance' with the necessary calculations
team_away_performance = football_data_cleaned.groupby('AWAY').agg({
    'AWAY_SCORE': 'sum',  # Total goals scored away
    'HOME_SCORE': 'sum',  # Total goals conceded away
    'AWAY_WIN': 'sum',    # Total away wins
    'AWAY_DRAW': 'sum'    # Total away draws
}).reset_index()

# Calculate the total points won away
team_away_performance['AWAY_POINTS'] = 3 * team_away_performance['AWAY_WIN'] + team_away_performance['AWAY_DRAW']

# Now we can recreate the bar charts with totals displayed above the bars
# Bar chart for the number of goals scored in away games per team with totals
plt.figure(figsize=(14, 6))
ax_away_scored = sns.barplot(x='AWAY', y='AWAY_SCORE', data=team_away_performance.sort_values('AWAY_SCORE', ascending=False))
add_totals_to_bars(ax_away_scored)
plt.title('Number of Goals Scored in Away Games per Team')
plt.xlabel('Team')
plt.ylabel('Goals Scored Away')
plt.xticks(rotation=90)
plt.show()

# Bar chart for the number of goals conceded in away games per team with totals
plt.figure(figsize=(14, 6))
ax_away_conceded = sns.barplot(x='AWAY', y='HOME_SCORE', data=team_away_performance.sort_values('HOME_SCORE'))
add_totals_to_bars(ax_away_conceded)
plt.title('Number of Goals Conceded in Away Games per Team')
plt.xlabel('Team')
plt.ylabel('Goals Conceded Away')
plt.xticks(rotation=90)
plt.show()

# Bar chart for total points scored in away games per team with totals
plt.figure(figsize=(14, 6))
ax_away_points = sns.barplot(x='AWAY', y='AWAY_POINTS', data=team_away_performance.sort_values('AWAY_POINTS', ascending=False))
add_totals_to_bars(ax_away_points)
plt.title('Total Points Scored in Away Games per Team')
plt.xlabel('Team')
plt.ylabel('Total Away Points')
plt.xticks(rotation=90)
plt.show()
```
